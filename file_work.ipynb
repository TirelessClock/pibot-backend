```python
import nltk
nltk.download('punkt')

def split_into_sentences(text):
    sentences = nltk.sent_tokenize(text)
    return sentences
```

```python
from docx import Document
import os

def checkLength(line):
    count = 0
    for c in line:
        if c == '\n' or c == ' ' or c == '\xa0':
            count += 1

    return (count != len(line))

def docWork(path):
    doc = Document(path)
    title = os.path.basename(path)[:-5]

    path_parts = path.split('\\')
    month = path_parts[-2]
    year = path_parts[-4]

    res = []
    for para in doc.paragraphs:
        if "***" in para.text:
            break
        if checkLength(para.text):
            content = para.text
            sentences = split_into_sentences(content)

            for sentence in sentences:
                res.append({"content": sentence, "tags": {}, "month": month, "year": year,})

    return res
```

```python
def entity_extraction(text):
    doc = nlp(text)
    labels_to_remove = ["CARDINAL", "PERCENT", "ORDINAL", "DATE", "QUANTITY", "TIME"]
    entities = []
    for ent in doc.ents:
        label = ent.label_
        if label not in labels_to_remove:
            entities.append(ent.text)

    entities = list(set(entities))

    return entities
```

```python
for item in database:
    item["tags"] = entity_extraction(item["content"])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def sort_documents_by_tfidf(indices, documents):
    all_documents = ' '.join(documents)
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([all_documents])
    feature_names = vectorizer.get_feature_names_out()
    tfidf_scores = dict(zip(feature_names, tfidf_matrix.toarray()[0]))
    indices_map = {idx: documents.index(database[idx]['content']) for idx in indices}
    sorted_indices = sorted(indices, key=lambda idx: max(tfidf_scores.get(word, 0) for word in documents[indices_map[idx]].split()), reverse=True)

    return sorted_indices
```

```python
sorted_tags = {}
for tag, indices in tag_doc.items():
    sorted_indices = sort_documents_by_tfidf(indices, [database[idx]['content'] for idx in indices])
    sorted_tags[tag] = sorted_indices
```

```python
def pick_docs(entities, st, k):
    res = {}
    for ent in entities:
        if ("modi" in ent.lower()) or ("prime minister" in ent.lower()) or ("pm" in ent.lower()) or ("narendra" in ent.lower()):
            continue
        if ent in st:
            indices = st[ent][:k]
            res[ent] = indices
    return res
```

```python
prompt = input("bolo: ")
entities = entity_extraction(prompt)

k = 3
doc_idx = pick_docs(entities, sorted_tags, k)
```

```python
from openai import OpenAI

with open('key.txt', 'r') as file:
    key = file.read()
    
client = OpenAI(api_key=key,)

def chatGPTinteraction(prompt):
    stream = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        stream=True,
    )

    res = ""
    for chunk in stream:
        if chunk.choices[0].delta.content is not None:
            res += chunk.choices[0].delta.content

    return res
```

```python
def promptEngineering(question, referenceDocIndices, entities):

    print(entities)
    
    if len(entities) == 0:
        return "Say 'Sorry, I can't answer this question based on the data I have. May I help you with something else?'"

    text = "PM said in his speech:\n"
    for tag, _ in entities.items():
        text += f"{tag}: "
        for idx in entities[tag]:
            text += f"{referenceDocIndices[idx]['content']}, "
        text += '\n'

    text += f"{prompt}. Answer starting with 'Based on his past speeches'. Respond briefly."

    return text
```
